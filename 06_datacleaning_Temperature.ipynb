```python
# Start by importing required modules
import pandas as pd

%matplotlib inline
```

```python
# Import the required data
# A change in encoding type was needed due to an error in data loading for food file.
temperature_df = pd.read_csv("Environment_Temperature_change_E_All_Data_NOFLAG.csv",  encoding='latin-1')
```

```python
# Have a quick look at the food dataframe
print(temperature_df.columns.is_unique)
print(temperature_df.shape)
print(temperature_df[temperature_df.duplicated() == True])
temperature_df.head()
```

```python
# Remove columns that are not required
temperature_df = temperature_df.drop(
    ['Area Code', 'Months Code', 'Element Code', 'Unit', 
    ],
    axis=1 
)
```

```python
temperature_df.head()
```

```python
# Rename the some of the column name
# Rename the remaining columns
temperature_df.rename(columns=lambda x: x[-4:] if x[0] == 'Y' else x, inplace=True)
temperature_df = temperature_df.rename(
    columns = {
        'Element': "Type", 'Area': 'Country'
        } 
)
temperature_df= temperature_df.set_index('Country')
temperature_df.head()
```

```python
# Fill the missing values with 0
temperature_df.fillna(value=0, inplace=True)
temperature_df.head()
```

```python
## Delete region rows.  We want only countries.
temperature_df.drop(
    index=['Africa','Americas', 'Annex I countries', 'Asia', 'Australia and New Zealand','Belgium-Luxembourg', 
           'Caribbean', 'Central America', 'Central Asia', 'China, Hong Kong SAR', 
           'China, Macao SAR', 'China, mainland', 'China, Taiwan Province of', 
           'Eastern Africa', 'Eastern Asia', 'Eastern Europe', 'Ethiopia PDR',
           'Europe', 'European Union (27)', 'European Union (28)', 'Land Locked Developing Countries',
           'Least Developed Countries', 'Low Income Food Deficit Countries', 'Melanesia', 'Micronesia',
           'Middle Africa', 'Net Food Importing Developing Countries', 'Non-Annex I countries', 'Northern Africa', 'Northern America', 
           'Northern Europe', 'Oceania', 'OECD', 'Polynesia', 'Serbia and Montenegro', 'Small Island Developing States', 
           'South America', 'South-eastern Asia', 'Southern Africa', 'Southern Asia', 'Southern Europe', 'Sudan (former)', 
           'USSR', 'Western Africa', 'Western Asia', 'Western Europe', 'World',
            ], axis=0, inplace=True) 
temperature_df
```

```python
print(temperature_df.index.value_counts())
```

```python
# Drop everything except month = 'Meteorological year' and Element = 'Temperature change'
temperature_df = temperature_df.loc[
    (temperature_df['Months'] == 'Meteorological year') ]
temperature_df
```

```python
## Delete certain temperature type.
temperature_df = temperature_df[temperature_df.Type != 'Standard Deviation']
temperature_df 
```

```python
# Remove columns that are not required
temperature_df = temperature_df.drop(
    ['Months', 'Type'],
    axis=1 
)
temperature_df
```

```python
# Output to csv file.
temperature_df.to_csv('Temperature.csv')
```
