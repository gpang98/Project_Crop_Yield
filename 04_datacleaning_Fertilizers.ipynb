```python
# Start by importing required modules
import os
import pandas as pd

%matplotlib inline
```

```python
# Define the directory path where the CSV file is located
directory_path = "Data_Input"

# Define the CSV file name
file_name = "Inputs_FertilizersNutrient_E_All_Data_NOFLAG.csv"
```

```python
# Construct the full file path by joining the directory path and file name
file_path = os.path.join(directory_path, file_name)
```

```python
# Import the required data
# A change in encoding type was needed due to an error in data loading for food file.
fertilizers_df = pd.read_csv(file_path,  encoding='latin-1')
```

```python
# Have a quick look at the food dataframe
print(fertilizers_df.columns.is_unique)
print(fertilizers_df.shape)
print(fertilizers_df[fertilizers_df.duplicated() == True])
fertilizers_df.head()
```

```python
# Delete unwanted columns
fertilizers_df.drop(
    columns=['Area Code', 'Item Code', 'Element Code', 'Unit',], axis=1, inplace=True)
```

```python
fertilizers_df
```

```python
# Rename the some of the column name
# Rename the remaining columns
fertilizers_df.rename(columns=lambda x: x[-4:] if x[0] == 'Y' else x, inplace=True)
fertilizers_df = fertilizers_df.rename(
    columns = {
        'Area': 'Country', 'Item': 'Fertilizer'
        } 
)
fertilizers_df = fertilizers_df.set_index('Country')
fertilizers_df.head()
#crop_df.info()
```

```python
# Fill the missing values with 0
fertilizers_df.fillna(value=0, inplace=True)
fertilizers_df.head()
```

```python
## Delete region rows.
fertilizers_df.drop(
    index=['Africa','Americas', 'Asia', 'Australia and New Zealand','Belgium-Luxembourg', 
           'Caribbean', 'Central America', 'Central Asia', 'China, mainland', 'China, Taiwan Province of', 
           'Eastern Africa', 'Eastern Asia', 'Eastern Europe', 'Ethiopia PDR',
           'Europe', 'European Union (27)', 'European Union (28)', 'Land Locked Developing Countries',
           'Least Developed Countries', 'Low Income Food Deficit Countries', 'Melanesia', 'Micronesia',  
           'Middle Africa', 'Net Food Importing Developing Countries', 'Northern Africa', 'Northern America', 
           'Northern Europe', 'Oceania', 'Polynesia', 'Serbia and Montenegro', 'Small Island Developing States', 
           'South America', 'South-eastern Asia', 'Southern Africa', 'Southern Asia', 'Southern Europe', 'Sudan (former)', 
           'USSR', 'Western Africa', 'Western Asia', 'Western Europe', 'World',
            ], axis=0, inplace=True) 
fertilizers_df
```

```python
fertilizers_df = fertilizers_df.loc[
    (fertilizers_df['Element'] == 'Agricultural Use') ]
```

```python
# Delete unwanted columns
fertilizers_df_copy = fertilizers_df.copy()
fertilizers_df_copy = fertilizers_df_copy.drop(columns=['Element'], axis=1)
```

```python
fertilizers_df_copy.head()
```

```python
fertilizers_df = fertilizers_df_copy.copy()

# Define the replacement dictionary
replacement_dict = {
    'Nutrient nitrogen N (total)': 'Nitrogen (N)',
    'Nutrient phosphate P2O5 (total)': 'Phosphate (P)',
    'Nutrient potash K2O (total)': 'Potash (K)'
}

# Replace values in the DataFrame using the dictionary
fertilizers_df.replace(to_replace=replacement_dict, inplace=True)

fertilizers_df
```

```python
# Specify directory to output csv file
output_directory_path = "Data_Output"
file_name = "Fertilizers.csv"
file_path = os.path.join(output_directory_path, file_name)
```

```python
# Output to csv file.
fertilizers_df.to_csv(file_path)
```
