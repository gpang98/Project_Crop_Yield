```python
# Start by importing required modules
import os
import pandas as pd

%matplotlib inline
```

```python
# Define the directory path where the CSV file is located
directory_path = "Data_Input"

# Define the CSV file name
file_name1 = "Worldbank_Rainfall_1961_1990_ed.csv"
file_name2 = "Worldbank_Rainfall_1991_2020_ed.csv"
```

```python
# Construct the full file path by joining the directory path and file name
file_path1 = os.path.join(directory_path, file_name1)
file_path2 = os.path.join(directory_path, file_name2)
```

```python
# Import the required data
# A change in encoding type was needed due to an error in data loading for food file.
#rain1_df = pd.read_csv("Worldbank_Rainfall_1961_1990.csv", index_col='Country')
#rain2_df = pd.read_csv("Worldbank_Rainfall_1991_2020.csv", index_col='Country')
rain1_df = pd.read_csv(file_path1)
rain2_df = pd.read_csv(file_path2)
```

```python
# Have a quick look at the food dataframe
print(rain1_df.columns.is_unique)
print(rain1_df.shape)
print(rain1_df[rain1_df.duplicated() == True])
rain1_df.head()
```

```python
print(rain2_df.columns.is_unique)
print(rain2_df.shape)
print(rain2_df[rain2_df.duplicated() == True])
rain2_df.head()
```

```python
# Fill the missing values with 0
rain1_df.fillna(value=0, inplace=True)

rain1_df
```

```python
# Strip leading and trailing spaces from column names
rain1_df = rain1_df.rename(columns={col: col.strip() for col in rain1_df.columns})

# Strip spaces from the "Country" column
rain1_df['Country'] = rain1_df['Country'].str.strip()
```

```python
rain1_df
```

```python
rain2_df.fillna(value=0, inplace=True)

rain2_df
```

```python
# Strip leading and trailing spaces from column names
rain2_df = rain2_df.rename(columns={col: col.strip() for col in rain2_df.columns})

# Strip spaces from the "Country" column
rain2_df['Country'] = rain2_df['Country'].str.strip()
```

```python
rain2_df
```

```python
result1_df = rain1_df.groupby(['Country', 'ISO3', 'Year']).mean()
result1_df = result1_df.unstack().reset_index()
```

```python
result2_df = rain2_df.groupby(['Country', 'ISO3', 'Year']).mean()
result2_df = result2_df.unstack().reset_index()
```

```python
result1_df
```

```python
result2_df
```

```python
rain_df = pd.merge(result1_df, result2_df, how='outer')

rain_df
```

```python
rain_df.columns = ['Country', 'Country Code', 1961, 1962, 1963, 1964, 1965, 1966, 1967, 1968, 1969,
                  1970, 1971, 1972, 1973, 1974, 1975, 1976, 1977, 1978, 1979,
                  1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989,
                  1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999,
                  2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009,
                  2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019,
                  2020,]
```

```python
rain_df = rain_df.set_index('Country')
```

```python
rain_df
```

```python
# Specify directory to output csv file
output_directory_path = "Data_Output"
file_name = "Rain.csv"
file_path = os.path.join(output_directory_path, file_name)
```

```python
# Output to csv file.
rain_df.to_csv(file_path)
```
