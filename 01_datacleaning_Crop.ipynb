```python
# Start by importing required modules
import pandas as pd

%matplotlib inline
```

```python
# Import the required data
# A change in encoding type was needed due to an error in data loading for food file.
crop_df = pd.read_csv("Production_Crops_E_All_Data_NOFLAG.csv",  encoding='latin-1')
```

```python
# Have a quick look at the food dataframe
print(crop_df.columns.is_unique)
print(crop_df.shape)
print(crop_df[crop_df.duplicated() == True])
crop_df.head()
```

```python
# Remove columns that are not required
crop_df = crop_df.drop(
    ['Area Code', 'Item Code', 'Unit', 'Element Code',
    ],
    axis=1 
)
```

```python
crop_df.head()
```

```python
# Rename the some of the column name
# Rename the remaining columns
crop_df.rename(columns=lambda x: x[-4:] if x[0] == 'Y' else x, inplace=True)
crop_df = crop_df.rename(
    columns = {
        'Item': 'Crop', 'Element': "Type", 'Area': 'Country'
        } 
)
crop_df= crop_df.set_index('Country')
crop_df.head()
#crop_df.info()
```

```python
# Fill the missing values with 0
crop_df.fillna(value=0, inplace=True)
crop_df.head()
print(crop_df['Crop'].value_counts())
```

```python
## Delete region rows.  We want only countries.
crop_df.drop(
    index=['Africa','Americas', 'Asia', 'Australia and New Zealand','Belgium-Luxembourg', 
           'Caribbean', 'Central America', 'Central Asia', 'China, Hong Kong SAR', 
           'China, Macao SAR', 'China, mainland', 'China, Taiwan Province of', 
           'Eastern Africa', 'Eastern Asia', 'Eastern Europe', 'Ethiopia PDR',
           'Europe', 'European Union (27)', 'European Union (28)', 'Faroe Islands', 'French Polynesia', 'Land Locked Developing Countries',
           'Least Developed Countries', 'Low Income Food Deficit Countries', 'Melanesia', 'Micronesia',
           'Middle Africa', 'Net Food Importing Developing Countries', 'New Caledonia', 'Niue', 'Northern Africa', 'Northern America', 
           'Northern Europe', 'Oceania', 'Polynesia', 'Serbia and Montenegro', 'Small Island Developing States', 
           'South America', 'South-eastern Asia', 'Southern Africa', 'Southern Asia', 'Southern Europe', 'Sudan (former)', 
           'USSR', 'Western Africa', 'Western Asia', 'Western Europe', 'World',
            ], axis=0, inplace=True) 
crop_df
```

# Drop the following Items:
Oilcrops-> include too many items. <br>
Sugar Crops Primary -> include sugar cane and sweeteners and derived products.<br>
Fruit Primary -> include almost all fruits.<br>
Rice, paddy (rice milled equivalent).

```python
## Delete certain crop type.
crop_df = crop_df[crop_df.Crop != 'Oilcrops']
crop_df = crop_df[crop_df.Crop != 'Sugar Crops Primary']
crop_df = crop_df[crop_df.Crop != 'Fruit Primary']
crop_df = crop_df[crop_df.Crop != 'Rice, paddy (rice milled equivalent)']
crop_df
```

```python
#Drop any yield less than 15000 hg/ha (so that Wheat is included)
crop_yield_df = crop_df[(crop_df.Type == 'Yield') & (crop_df['2019'] >= 15000)].reset_index()
crop_yield_df
```

```python
#Drop any production less than 15000 tonnes (so that Wheat is included)
crop_production_df = crop_df[(crop_df.Type == 'Production') & (crop_df['2019'] >= 15000)].reset_index()
crop_production_df
```

```python
#Drop any area harvest less than 100 ha (so that Wheat is included)
crop_area_df = crop_df[(crop_df.Type == 'Area harvested') & (crop_df['2019'] >= 100)].reset_index()
crop_area_df
```

```python
#Merging all the dataframe together
crop_df = pd.concat([crop_yield_df, crop_production_df, crop_area_df], axis=0).reset_index()
del crop_df['index']
crop_df
```
