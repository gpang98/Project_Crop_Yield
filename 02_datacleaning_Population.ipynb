```python
# Start by importing required modules
import os
import pandas as pd

%matplotlib inline
```

```python
# Define the directory path where the CSV file is located
directory_path = "Data_Input"

# Define the CSV file name
file_name = "Population_E_All_Data_NOFLAG.csv"
```

```python
# Construct the full file path by joining the directory path and file name
file_path = os.path.join(directory_path, file_name)
```

```python
# Import the required data
# A change in encoding type was needed due to an error in data loading for food file.
population_df = pd.read_csv(file_path,  encoding='latin-1')
```

```python
# Have a quick look at the food dataframe
print(population_df.columns.is_unique)
print(population_df.shape)
print(population_df[population_df.duplicated() == True])
population_df.head()
```

```python
# Delete unwanted columns Y2021-2100
population_df = population_df.loc[:,'Area':'Y2021']
population_df
```

```python
# Further delete unwanted columns
population_df.drop(
    columns=['Item Code', 'Item', 'Element Code', 'Unit', 'Y1950', 'Y1951', 'Y1952', 'Y1953', 'Y1954', 
             'Y1955', 'Y1956', 'Y1957', 'Y1958', 'Y1959', 'Y1960',], axis=1, inplace=True)
```

```python
population_df
```

```python
# Rename the some of the column name
# Rename the remaining columns
population_df.rename(columns=lambda x: x[-4:] if x[0] == 'Y' else x, inplace=True)
population_df = population_df.rename(
    columns = {
        'Element': "Type", 'Area': 'Country'
        } 
)
population_df = population_df.set_index('Country')
population_df.head()
#crop_df.info()
```

```python
# Fill the missing values with 0
population_df.fillna(value=0, inplace=True)
population_df.head()
```

```python
## Delete region rows.
population_df.drop(
    index=['Africa','Americas', 'Asia', 'Australia and New Zealand', 'Belgium-Luxembourg', 
           'Caribbean', 'Central America', 'Central Asia', 'China, Hong Kong SAR', 
           'China, Macao SAR', 'China, mainland', 'China, Taiwan Province of', 
           'Eastern Africa', 'Eastern Asia', 'Eastern Europe', 'Ethiopia PDR',
           'Europe', 'European Union', 'Land Locked Developing Countries',
           'Least Developed Countries', 'Low Income Food Deficit Countries', 'Melanesia', 'Micronesia',
           'Middle Africa', 'Net Food Importing Developing Countries', 'Northern Africa', 'Northern America', 
           'Northern Europe', 'Oceania', 'Pacific Islands Trust Territory', 'Polynesia', 'Serbia and Montenegro', 'Small Island Developing States', 
           'South America', 'South-Eastern Asia', 'Southern Africa', 'Southern Asia', 'Southern Europe', 'Sudan (former)', 
           'USSR', 'Western Africa', 'Western Asia', 'Western Europe', 'World',
            ], axis=0, inplace=True) 
population_df
```

```python
## Delete certain crop type.
population_df = population_df[population_df.Type != 'Total Population - Male']
population_df = population_df[population_df.Type != 'Total Population - Female']
population_df = population_df[population_df.Type != 'Rural population']
population_df = population_df[population_df.Type != 'Urban population']

# Delete unwanted columns
population_df.drop(
    columns=['Type'], axis=1, inplace=True)

population_df
```

```python
print(population_df.index.value_counts())
```

```python
# Specify directory to output csv file
output_directory_path = "Data_Output"
file_name = "Population.csv"
file_path = os.path.join(output_directory_path, file_name)
```

```python
# Output to csv file.
population_df.to_csv(file_path)
```
